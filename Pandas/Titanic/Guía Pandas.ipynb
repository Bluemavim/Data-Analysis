```python
import pandas as pd
import numpy as np

df= pd.read_csv("titanic3.csv")
df
```

```python
#2)Obtener los nombres de todas las columnas. 
print(f"Nombres de las columnas: {df.columns}")
```

```python
#Luego, iterar ese objeto columna a columna e indicar valores restantes.
for col in df.columns:
    print(f"{col.upper()} tiene {df[col].isnull().sum()} valores faltantes.")
```

```python
#3) Averiguar cuál fue la media de las edades. Imprimir el valor
media_edad = df["age"].mean()
mediana_edad = df["age"].median()
moda_edad = df["age"].mode()
varianza_edad = df["age"].var()
desvio_edad = df["age"].std()

print(f"El promedio de edad de lxs pasajerxs fue: {media_edad:.3f}.")
print(f"La mediana de la edad lxs pasajerxs fue: {mediana_edad:.3f}.")
print(f"La moda de la edad lxs pasajerxs fue: {moda_edad}.")
print(f"La varianza de la edad lxs pasajerxs fue: {varianza_edad}.")
print(f"El desvío estándar la edad lxs pasajerxs fue: {desvio_edad}.")
```
