```python
import pandas as pd
import numpy as np

df= pd.read_csv("opsd_germany_daily.csv")
df
```

```python
#2) Rellenar con 0 los valores faltantes en Wind y Solar.
#Algunas formas de rellenar todos los valores
df = df.fillna(0)
df
```

```python
#Otra versión del anterior con función lambda. En comentado va reemplazar todo.
from pandas.api.types import is_numeric_dtype
df = df.apply(lambda x: x.fillna(0) if is_numeric_dtype(x) else x)
df

#df['Wind','Solar'].apply(lambda x: x.fillna(0) if is_numeric_dtype(x) else x) 

```

```python
#2) Rellenar con 0 los valores faltantes en Wind y Solar.
#Al usar "inplace" se modifica el dataframe original.
df.Wind.fillna(0, inplace= True)
df.Solar.fillna(0, inplace= True)
df
```

```python
#La columna Wind+Solar tiene que coindicir con la suma de Wind+Solar
df.eval("`Wind+Solar`= Wind + Solar", inplace=True)
df
```

```python
#4) Crear una tabla pivote que visualice el consumo de Wind+Solar según la fecha, debe ser vertical.
df.drop("Wind+Solar", axis=1, inplace=True)
df
```

```python
df.rename(columns={"BACKTICK_QUOTED_STRING_Wind_PLUS_Solar":"Wind+Solar"}, inplace=True)
df
```

```python
#4) Crear una tabla pivote que visualice el consumo de Wind+Solar según la fecha, debe ser vertical.

tabla_pivote =  df.pivot_table("Consumption", index="Date")
tabla_pivote
```

```python
#5) Reemplazar los valores faltantes de la tabla pivote por la media.
tabla_pivote.fillna(df.Consumption.mean(), inplace=True)
tabla_pivote
```
